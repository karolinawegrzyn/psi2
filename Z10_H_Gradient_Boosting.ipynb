```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions
```

# Gradient Boosting

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

ada_clf = GradientBoostingClassifier(n_estimators=1, learning_rate=0.5, random_state=42)
ada_clf.fit(X_train, y_train)
```

```python
plot_decision_regions(X_train, y_train, ada_clf)
plt.show()
```

# Zad.

Sprawd rne parametry dla GradientBoostingClassifier

* n_estimators=2, learning_rate=0.5
* n_estimators=2, learning_rate=1
* n_estimators=10, learning_rate=0.5
* n_estimators=10, learning_rate=1

```python
ada_clf1 = GradientBoostingClassifier(n_estimators=2, learning_rate=0.5, random_state=42)
ada_clf1.fit(X_train, y_train)

ada_clf2 = GradientBoostingClassifier(n_estimators=2, learning_rate=1, random_state=42)
ada_clf2.fit(X_train, y_train)

ada_clf3 = GradientBoostingClassifier(n_estimators=10, learning_rate=0.5, random_state=42)
ada_clf3.fit(X_train, y_train)

ada_clf4 = GradientBoostingClassifier(n_estimators=10, learning_rate=1, random_state=42)
ada_clf4.fit(X_train, y_train)


ada_clfs = [ada_clf1, ada_clf2, ada_clf3, ada_clf4]

for ada_clf in ada_clfs:
  plot_decision_regions(X, y, ada_clf)
  plt.show()
```
