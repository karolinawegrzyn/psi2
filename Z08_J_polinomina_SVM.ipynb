```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics

from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

# Linear SVM Classification

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=.1, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

# Polynomial Kernel

# Zad.

Naucz SVM z jÄ…drem "poly" o parametrach

* degree=1, coef0=0, C=1
* degree=2, coef0=0, C=1
* degree=3, coef0=0, C=1
* degree=3, coef0=1, C=1

```python
degree = [1,2,3,3]
coef0 = [0,0,0,1]

for i in range (0, 4):
  svc = SVC(kernel="poly", degree = degree[i], coef0 = coef0[i], C = 1)
  svc.fit(X,y)
  plot_decision_regions(X,y,svc)
  plt.show() 
```
