```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics

from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

# RBF SVM Classification

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=.1, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

# Kernel trick

# Zad.

Naucz SVM z jądrem "rbf" o parametrach

* gamma=0.01, C=1.0
* gamma=0.1, C=1.0
* gamma=1, C=1.0
* gamma=10, C=1.0
* gamma=100, C=1.0
* gamma=100, C=10
* gamma=100, C=100

```python
gamma = [0.01,0.1,1,10,100,100,100]
c = [1,1,1,1,1,10,100]

for i in range (0, 7):
  svc = SVC(kernel="rbf", gamma = gamma[i], C = c[i])
  svc.fit(X,y)
  plot_decision_regions(X,y,svc)
  plt.show() 
```
